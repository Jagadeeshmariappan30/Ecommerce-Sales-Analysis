```python
df = pd.read_csv('ecommerce_sales_dataset.csv')
```

```python
df.isnull().sum()
```

```python
df['rating'] = df.groupby('category')['rating'].transform(lambda x: x.fillna(x.median()))
```

```python
df['discount_percentage'] = df.groupby('category')['discount_percentage'].transform(lambda x: x.fillna(x.median()))
```

```python
df['payment_method'] = df['payment_method'].fillna(df['payment_method'].mode()[0])
```

```python
df['city'] = df['city'].fillna(df['city'].mode()[0])
```

```python
df.isnull().sum()
```

```python
df.duplicated().sum()
```

```python
df.dtypes
```

```python
df['order_date'] = pd.to_datetime(df['order_date'],format='%d-%m-%Y')
```

```python
df['order_date'].info()
```

```python
df.columns
```

Feature Engineering

```python
df['discount_amount'] = df['actual_price'] - df['discounted_price']
df['discount_amount'].head(5)
```

```python
df['order_month'] = df['order_date'].dt.month

df['order_month'].head(5)
```

```python
df['order_year'] = df['order_date'].dt.year

df['order_year'].head(5)
```
